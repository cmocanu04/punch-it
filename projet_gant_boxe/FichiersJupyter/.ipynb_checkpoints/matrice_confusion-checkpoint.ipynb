```python
import joblib
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Définir les noms des fichiers ---
IMPUTER_FILE = '../joblib/Commun/imputer_gant.joblib'
SCALER_FILE = '../joblib/Commun/scaler_gant.joblib'
KNN_MODEL_FILE = '../joblib/Commun/knn_model_gant.joblib'
TEST_DATA_FILE = '../joblib/Eldin/dataset_entrainement_toutes_sessions.csv' # Le fichier CSV de test

# --- 2. Charger les modèles et le préprocesseur entraînés ---
print(f"Chargement de l'imputer depuis '{IMPUTER_FILE}'...")
imputer = joblib.load(IMPUTER_FILE)
print(f"Chargement du scaler depuis '{SCALER_FILE}'...")
scaler = joblib.load(SCALER_FILE)
print(f"Chargement du modèle KNN depuis '{KNN_MODEL_FILE}'...")
knn_model = joblib.load(KNN_MODEL_FILE)

# --- 3. Charger les données de test ---
print(f"Chargement des données de test depuis '{TEST_DATA_FILE}'...")
try:
    test_df = pd.read_csv(TEST_DATA_FILE)
    print("Données de test chargées avec succès.")
    print("Aperçu des données de test:")
    print(test_df.head())
    print("\nInformations sur les données de test:")
    test_df.info()
except FileNotFoundError:
    print(f"Erreur : Le fichier '{TEST_DATA_FILE}' n'a pas été trouvé.")
    exit() # Quitte le script si le fichier n'est pas trouvé
except Exception as e:
    print(f"Une erreur est survenue lors du chargement du fichier CSV : {e}")
    exit()

# --- 4. Séparer les caractéristiques (X_test) et la cible (y_test) ---
# Définir les noms de colonnes des caractéristiques et de la cible
feature_columns = [
    'Min_AccX_Val', 'Max_AccX_Val', 'AccX_Amplitude', 'AccX_Ord(1=Max>Min)',
    'Min_AccZ_Val', 'Max_Roll_Val', 'Min_Roll_Val', 'Roll_Ord(1=Max>Min)',
    'Max_Pitch_Val', 'Min_Pitch_Val', 'Pitch_Ord(1=Max>Min)'
]
target_column = 'Type_Coup'

# Vérifier si toutes les colonnes nécessaires sont présentes dans le DataFrame
missing_features = [col for col in feature_columns if col not in test_df.columns]
if missing_features:
    print(f"Erreur : Colonnes de caractéristiques manquantes dans le fichier CSV : {missing_features}")
    exit()
if target_column not in test_df.columns:
    print(f"Erreur : Colonne cible '{target_column}' manquante dans le fichier CSV.")
    exit()

X_test = test_df[feature_columns]
y_test = test_df[target_column]

print("\nPréparation des données de test:")
print(f"Forme de X_test: {X_test.shape}")
print(f"Forme de y_test: {y_test.shape}")

# --- 5. Appliquer l'imputer aux données de test ---
# L'imputer est appliqué uniquement sur les colonnes des features
print("Application de l'imputer sur X_test...")
# X_test_imputed sera un tableau numpy, pas un DataFrame
X_test_imputed = imputer.transform(X_test)

# --- 6. Appliquer le scaler aux données de test imputées ---
print("Application du scaler sur X_test imputé...")
# X_test_scaled sera également un tableau numpy
X_test_scaled = scaler.transform(X_test_imputed)

# --- 7. Faire les prédictions avec le modèle KNN ---
print("Réalisation des prédictions sur X_test mis à l'échelle...")
y_pred = knn_model.predict(X_test_scaled)

print("\nPrédictions terminées.")
print(f"Aperçu des vraies étiquettes (y_test):\n{y_test.head()}")
print(f"\nAperçu des prédictions (y_pred):\n{y_pred[:5]}")

# --- 8. Calculer la matrice de confusion ---
print("\nCalcul de la matrice de confusion...")
cm = confusion_matrix(y_test, y_pred)
print("Matrice de Confusion:\n", cm)

# --- 9. Visualiser la matrice de confusion ---
print("Génération de la visualisation de la matrice de confusion...")

# Obtenir les labels de classes uniques pour l'affichage.
# Il est préférable d'utiliser les classes connues du modèle (knn_model.classes_).
# Sinon, on utilise l'union des classes présentes dans y_test et y_pred.
if hasattr(knn_model, 'classes_'):
    class_labels = knn_model.classes_
else:
    class_labels = np.unique(np.concatenate((y_test, y_pred)))

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_labels)

fig, ax = plt.subplots(figsize=(10, 10))
disp.plot(cmap=plt.cm.Blues, ax=ax)
plt.title('Matrice de Confusion KNN')
plt.xlabel('Prédiction')
plt.ylabel('Vraie Valeur')
plt.tight_layout() # Ajuste les paramètres du tracé pour qu'il rentre dans la figure
plt.savefig('confusion_matrix_knn.png') # Sauvegarde la figure en tant qu'image PNG
print("Matrice de confusion sauvegardée sous 'confusion_matrix_knn.png'")

print("\nCode exécuté avec succès. La matrice de confusion a été calculée et sauvegardée.")
```
